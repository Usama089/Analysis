```python
# to plot multiple graph:

df[["Customer_Age","Dependent_count"]].hist(figsize=(15,5),color="blue")
plt.tight_layout() #automatically adjusts subplot parameters so that the subplot(s) fits in to the figure area.
plt.show()         #use to display the all the figures.

'''OBSERVATIONS:
Highest number of customers range between age 44-50.
Lowest number of customer range between age 67-73.

Dependent count 3 is high.
'''
```

```python
# Line-Plot: used to access plot in lines.

x = np.arange(10) # to plot number range till 10
y = np.sqrt(x)    # return the non-negative square-root of an array
sns.lineplot(x,y) # for plotting lineplot in x and y axis
plt.grid(True)    # grid function marks horizontal and vertical line on graph for better understanding.
plt.show()
```

```python
# dist-plot: used to access graph in line and histogram format.

sns.distplot(df["Customer_Age"])
plt.grid()
plt.show()

'''OBSERVATION:
Line on the upper side of the histogram shows density between 0.04-0.05.
'''
```

```python
# count-plot: countplot() method is used to Show the counts of observations in each categorical bin using bars.

sns.countplot(data=df,x='Gender',hue='Education_Level') # hue parameter take column name for colour encoding.
plt.grid()
plt.show()

'''OBSERVATION:
The highest education-level in both male and female is graduation.
High-school education-level in both male and female stands on 2nd number.
Lowest education-level is doctorate in both.
'''
```

```python
# sub-plot is used to divide and plot multiple columns on the graph.
 
plt.subplot(2,2,1) # plotting graph on position 1
df['Customer_Age'].value_counts().plot(figsize=(20,20),kind='bar',color='blue') # kind="bar" is used to plot bar type graph.
plt.title("Customer_Age Diagram") # plt.title used to give diagram heading
plt.ylabel("No. of Customers")    # ylabel is used to label no. of customers on y-axis.
plt.xlabel("Customer Age")        # xlabel is used to label customer age on x-axis.
plt.grid()

plt.subplot(2,2,2) # plotting graph on position 2
df['Gender'].value_counts().plot(figsize=(20,20),kind='bar',color='red')
plt.title("Customer_Gender Diagram")
plt.ylabel("No. of Customers")
plt.xlabel("Customer Gender")
plt.grid()

plt.subplot(2,2,3) # plotting graph on position 3
df['Income_Category'].value_counts().plot(figsize=(20,20),kind='bar',color='purple')
plt.title("Income_Category Diagram")
plt.ylabel("No. of Customers")
plt.xlabel("Income Category")
plt.grid()

plt.subplot(2,2,4) # plotting graph on position 4
df['Card_Category'].value_counts().plot(figsize=(20,20),kind='bar',color='yellow')
plt.title("Card_Category Diagram")
plt.ylabel("No. of Customers")
plt.xlabel("Card Category")
plt.grid()

plt.tight_layout()
plt.show()

'''OBSERVATION:
1.Customer_Age Diagram:
  In the Customer_Age Diagram the highest number of customers are 200 and above with age 45.
  The customers with age 49 are on 2nd position.
  and customers with age 47 on third.
  Lowest number of customers is of age 73
  
2.Customer_Gender Diagram:
  The number of male customers are more then female customers.
  
3.Income_Category Diagram:
  1400+ customers have income range less then 40000$.
  Customers between 400-420 have the highest income range of 120k$.
  
4.Card_Category Diagram:
  4500+ Customers have Blue card category.
  10 and below customers own platinum card.
'''
```

```python
#Pie-chart: is a circular representation on the data

explode=[0,0.1,0.1,0.3]  #explode parameter is used to highlight chart size. here, 0.1 is used to highlight some columns.
df['Card_Category'].value_counts().plot(figsize=(8,8),kind='pie',autopct='%.2f%%',explode=explode)
plt.title("Card-Category Pie-Chart")  #autopct="% %" is used to access pie area.
plt.xlabel("Type of Cards")
plt.ylabel("Card Per")

'''
By using pieplot we can clearly see that the number of blue card customers is on 94.54%.
and the lowest is 0.06% which is platinum cards.'''
```

```python
df.corr() #shows co-relation between different columns.
```

```python
# heatmap data visualization is a method of graphically representing numerical data where the value of each data point is indicated using colors.

plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True,cmap='Blues') # annot=True to access values in each cell, cmap is used to give color code.
plt.show()
```

```python
'''Box-plot or Whiskers-plot:
A boxplot is a standardized way of displaying the distribution of data based on a five number summary 
(“minimum”, first quartile [Q1], median, third quartile [Q3] and “maximum”). 
It can tell you about your outliers and what their values are.
'''

plt.figure(figsize=(20,20))
sns.boxplot(data=df,x='Income_Category',y='Customer_Age')
plt.show()

#There are no out-liers in the data-set.
```

```python
#scatterplot : A scatterplot is a type of data display that shows the relationship between two numerical variables in dot.

plt.figure(figsize=(8,8))
sns.scatterplot(data=df,x='Education_Level',y='Income_Category')
plt.grid()
plt.show()

'''OBSERVATION:
here, income category is equally distributed with education level
'''
```

```python
#Pair-plot: The Seaborn Pairplot allows us to plot pairwise relationships between variables within a dataset.

plt.figure(figsize=(10,10))
sns.pairplot(df[['Customer_Age','Credit_Limit','Total_Revolving_Bal']],diag_kind='hist')
plt.show()
```
